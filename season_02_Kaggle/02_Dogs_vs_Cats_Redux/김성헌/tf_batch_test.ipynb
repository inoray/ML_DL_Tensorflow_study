```python
import tensorflow as tf
import cv2
import os
import numpy as np
import re
```

```python
DATA_DIR = "../data/"
TRAIN_DIR = DATA_DIR + "train_resize/"

train_dogs =   [TRAIN_DIR+i for i in os.listdir(TRAIN_DIR) if 'dog' in i]
train_dogs.sort(key=lambda var:[int(x) if x.isdigit() else x for x in re.findall(r'[^0-9]|[0-9]+', var)])
train_cats =   [TRAIN_DIR+i for i in os.listdir(TRAIN_DIR) if 'cat' in i]
train_cats.sort(key=lambda var:[int(x) if x.isdigit() else x for x in re.findall(r'[^0-9]|[0-9]+', var)])
train_dog_cat = train_dogs + train_cats

print("train_dogs: ", len(train_dogs))
print("train_cats: ", len(train_cats))
print("train_dog_cat: ", len(train_dog_cat))
```

```python
label_dog = [0 for i in range(len(train_dogs))]
label_cat = [1 for i in range(len(train_cats))]
label = label_dog + label_cat
label_one_hot = np.eye(2)[label]

print("label_one_hot: ", len(label_one_hot))
```

```python
train_dog_cat = train_dog_cat[:25]
label_one_hot = label_one_hot[:25]

input_queue = tf.train.slice_input_producer([train_dog_cat, label_one_hot], shuffle=False )

image_files = tf.read_file(input_queue[0])

image_list = tf.image.decode_jpeg(image_files, channels=3)
image_list.set_shape([150, 150, 3])

batch_size = 20
train_batch = tf.train.batch([input_queue[0], input_queue[1]], batch_size=batch_size)
```

```python
sess = tf.Session()

total_batch = int(len(train_dogs) / batch_size)

coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)

epoch = 2

for e in range(epoch):
    for i in range(total_batch):
        train_images, train_labels = sess.run(train_batch)
        print(e +1, i + 1)
        print(train_images, train_labels)
coord.request_stop()
coord.join(threads)
```
